## Final Results

```python
df_final = s1['sim_results'][['Date', 'Close', 'num_shares_bought', 'cash_to_invest']]
df_final = df_final.rename(columns={'num_shares_bought':'s1_num_shares_bought',
                                    'cash_to_invest': 's1_cash_to_invest'})
df_final = df_final.merge(s2['sim_results'][['Date', 'num_shares_bought', 'cash_to_invest']], on='Date')
df_final = df_final.rename(columns={'num_shares_bought':'s2_num_shares_bought',
                                    'cash_to_invest': 's2_cash_to_invest'})
df_final['year'] = df_final['Date'].dt.to_period('Y')
df_final
```

```python
df_final_g = df_final.groupby('year').agg({'s1_num_shares_bought':'sum',
                                           's2_num_shares_bought':'sum',
                                           'Close': 'last'}).reset_index()
df_final_g['s1_num_shares_bought_cumsum'] = np.cumsum(df_final_g['s1_num_shares_bought'])
df_final_g['s2_num_shares_bought_cumsum'] = np.cumsum(df_final_g['s2_num_shares_bought'])
df_final_g['s1_value'] = df_final_g['Close']*df_final_g['s1_num_shares_bought_cumsum']
df_final_g['s2_value'] = df_final_g['Close']*df_final_g['s2_num_shares_bought_cumsum']
df_final_g
```
